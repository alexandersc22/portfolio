### Analysis

##### What's the team with the most players in the league?

```python
df.info()
```

```python
df.groupby('Team')['Player'].size().sort_values(ascending=False).head()
```

##### Who are the top 5 highest-scoring players (pts) who have played at least 50 games (G), and what teams do they play for?

```python
df.loc[df['G']>=50, ['Player','Team','pts']].sort_values('pts',ascending=False).head(5)
```

##### Which team (Team) has the highest average Net Rating (net_rating) across all their rostered players?

```python
df.groupby('Team')['net_rating'].mean().sort_values(ascending=False).head(1)
```

##### Which colleges produced the most NBA players in this dataset?

```python
df['college'].value_counts().head()
```

##### Which players are the true offensive engines — defined as having a Usage Percentage (usg_pct) over 25% AND a True Shooting Percentage (ts_pct) over 60%?

```python
df.loc[(df['usg_pct']>0.25) & (df['ts_pct']>0.6), ['Player', 'usg_pct', 'ts_pct']].sort_values('usg_pct', ascending=False).head()
```

##### Which NBA teams rely the most on older, experienced players? Calculate the exact percentage of each team's roster that is made up of "Veterans" (players aged 30 or older).

```python
veteran_summary =df.groupby('Team').agg({
    'Age': lambda x: (x>=30).mean() * 100
})

veteran_summary = veteran_summary.rename(columns={'Age': 'veteran_pct'})
veteran_summary.sort_values('veteran_pct', ascending=False).head()
```
